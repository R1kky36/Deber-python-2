**Sistema de cobro con fecha de compra**

```python

class Cliente:
  def __init__(self, nombre, cedula, saldo=0.0, historial=[], array=[]):
    self.nombre = nombre
    self.cedula = cedula
    self.saldo = saldo
    self.historial = historial
    self.array = array

  def comprar(self, producto, fecha, ):
    if self.saldo >= producto.precio:
        self.saldo -= producto.precio
        self.historial= {"Producto":producto, "Fecha":fecha}
        self.array.append(self.historial)

    else:
      print("Ya no hay saldo suficiente")
```

```python
class Producto:
  def __init__(self, nombre, precio):
    self.nombre = nombre
    self.precio = precio
```

```python
c1 = Cliente(nombre="Juan", cedula="123", saldo=100)
```

```python
p1 = Producto(nombre="Café", precio=5)
```

```python
p2 = Producto(nombre="Helado", precio=20)
```

```python
p3 = Producto(nombre="Galletas", precio=15)
```

```python
c1.comprar(p1, "2025/10/21")
c1.comprar(p2, "2025/10/21")
c1.comprar(p3, "2025/10/21")
c1.comprar(p1, "2025/10/21")
c1.comprar(p2, "2025/10/21")
c1.comprar(p3, "2025/10/21")
c1.comprar(p1, "2025/10/21")
```
